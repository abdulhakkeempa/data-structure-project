```python
import pandas as pd
from shapely.geometry import Point
from geopandas import GeoDataFrame
import geopandas as gpd
import matplotlib.pyplot as plt


df = pd.read_csv("covid_19_clean_complete.csv")

geometry = [Point(xy) for xy in zip (df['Long'],df['Lat'])]
gdf = GeoDataFrame(df,geometry=geometry)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(15, 15)), marker='o', color='red', markersize=15);
```

```python
import math

def findpoints(lon, lat):
    radius = 1
    N = 1000 
    # generate points
    circlePoints = []
    for k in range(N):
        angle = math.pi*2*k/N
        dx = radius*math.cos(angle)
        dy = radius*math.sin(angle)
        point = {}
        point['lat']= lat + (180/math.pi)*(dy/6371) #Earth Radius
        point['lon']= lon + (180/math.pi)*(dx/6371)/math.cos(lon*math.pi/180) #Earth Radius
        # add to list
        circlePoints.append(point)

    return circlePoints


#edapally signal
latitude = 10.024637079537596
longitude = 76.30795060670427

sampleList = []
sampleList = findpoints(longitude,latitude)
for i in sampleList:
  print(i)
```
